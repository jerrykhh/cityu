```python
import pandas as pd 
import seaborn as sns

dataset = pd.read_csv("vertebral_column_data/column_3C.dat", header = None, sep = " ")
train_dataset = dataset[dataset.columns[:-1]]
training_classis = dataset.iloc[:,-1]
from sklearn import preprocessing 
le = preprocessing.LabelEncoder() 
dataset_target_labels = le.fit_transform(training_classis)
```

```python
from matplotlib import pyplot as plt 
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster 
from sklearn.cluster import KMeans 
```

```python
dataset.iloc[:,0]
```

```python
hcs = []
hcs_title = ['Hierarchical Clustering (single)', 'Hierarchical Clustering (complete)', 'Hierarchical Clustering (average)']
```

```python
hcs.append(linkage(train_dataset,method='single'))
hcs.append(linkage(train_dataset, method='complete'))
hcs.append(linkage(train_dataset, method='average'))
```

```python
index = 0
for hc in hcs:
    plt.figure(figsize=(25, 10))
    dendrogram(hc)
    plt.title(hcs_title[index])
    index+=1
    plt.show()
    
```

```python
for max_dis in range(20,210,30):
    for hc in hcs:
        clusters = fcluster(hc,max_dis, criterion='distance')
        plt.scatter(dataset.iloc[:,0],dataset.iloc[:,4], c=clusters)
        plt.show()
        print("max_dis:" + str(max_dis))
        print(clusters)
```

```python
for k in range(5,0,-1):
    for hc in hcs:
        clusters = fcluster(hc,k, criterion='maxclust')
        plt.scatter(dataset.iloc[:,5],dataset.iloc[:,4], c=clusters)
        plt.show()
        print("k:" + str(k))
        print(clusters)
```

```python
for n_c in range(5,0,-1):
    km = KMeans(n_clusters=n_c) 
    km.fit(train_dataset) 
    plt.scatter(dataset.iloc[:,0],dataset.iloc[:,4], c=km.labels_)
    plt.show()
    print("index" + str(n_c))
    print(km.labels_+1)
```
